```python
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/krishnaik06/Stock-Sentiment-Analysis/master/Data.csv', encoding = "ISO-8859-1")
```

```python
df.head(2)
```

```python
train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']
```

```python
#removing special characters
data =train.iloc[:, 2:27]
data.replace("[^a-zA-Z]", " ", regex =True, inplace = True)
```

```python
#renaming columnname for ease for access
list1 = [i for i in range(25)]
new_Index = [str(i) for i in list1]
data.columns= new_Index
data.head(5)
```

```python
# converting the healkines to lower case
for i in new_Index:
    data[i] = data[i].str.lower()
data.head(1)
```

```python
" ".join(str(x) for x in data.iloc[1, 0:25])
```

```python
#combinig all heaflines to one para
headlines = []
for row in range(0, len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row, 0:25]))
```

```python
headlines[0]
```

```python
#vetorizing the data 
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier 
```

```python
# creating bag of words model
countvector = CountVectorizer(ngram_range = (2, 2))
traindataset = countvector.fit_transform(headlines)
```

```python
# implamenting random forest
randomclassifier = RandomForestClassifier(n_estimators = 200, 
                                         criterion = "entropy")
randomclassifier.fit(traindataset, train["Label"])
```

```python
#predicting the test datset
test_tranform= []
for row in range(0, len(test.index)):
    test_tranform.append(" ".join(str(x) for x in test.iloc[row, 2:27]))
test_dataset = countvector.transform(test_tranform)
predictions = randomclassifier.predict(test_dataset)
```

```python
predictions
```

```python
from sklearn import metrics
import numpy as np
import itertools
import seaborn as sns 
```

```python
y_pred = predictions
y_test = test["Label"]
score  = metrics.accuracy_score(y_test, y_pred)
print("acuuracy score is {}".format(score))
cm = metrics.confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot = True, xticklabels = ["fake", "real"], 
            yticklabels = ["fake", "real"], cmap="YlGnBu")
```

```python
from sklearn.metrics import confusion_matrix
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=metrics.accuracy_score(test['Label'],predictions)
print(score)
report=metrics.classification_report(test['Label'],predictions)
print(report)
```
